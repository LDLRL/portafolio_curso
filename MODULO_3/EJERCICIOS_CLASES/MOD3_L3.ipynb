```python
import pandas as pd
import kagglehub
import os

##IMPORTANTE: SE USA CARGA DIRECTA ARCHIVOS .CSV Y .XLSX DESDE KAGGLE USANDO EL TOKEN DE MI CUENTA PERSONAL.
# EL CODIGO SE HA EJECUTADO SIN PROBLEMAS.
```

```python
print("=== 1.a. Cargar archivo CSV: Titanic ===")

# Descargar CSV desde Kaggle
path_csv = kagglehub.dataset_download("fossouodonald/titaniccsv")
csv_files = os.listdir(path_csv)
print("Archivos disponibles:", csv_files)

# Cargar el archivo CSV
df_csv = pd.read_csv(os.path.join(path_csv, "titanic.csv"))
print("\n▶ Primeras 5 filas del CSV (Titanic):")
display(df_csv.head())
```

```python
print("\n=== 1.b. Cargar archivo Excel: Liga Española ===")

# Ya habías descargado el archivo Excel antes, usamos ese mismo
path_excel = kagglehub.dataset_download("jbatistazerpa/liga-espaola-2023-2024-resultados")
excel_files = os.listdir(path_excel)
print("Archivos disponibles:", excel_files)

# Cargar el archivo Excel
df_excel = pd.read_excel(os.path.join(path_excel, "LigaEspanola2023-2024-Resultados.xlsx"))
print("\n▶ Primeras 5 filas del archivo Excel:")
display(df_excel.head())
```

```python
print("\n=== 1.c. Cargar tabla web: Wikipedia Población Mundial ===")

# Leer tabla desde Wikipedia
url_web = "https://en.wikipedia.org/wiki/List_of_countries_and_dependencies_by_population"
tablas = pd.read_html(url_web)

# Mostramos cuántas tablas se encontraron
print(f"▶ Se encontraron {len(tablas)} tablas.")

# Elegimos la primera tabla (generalmente es la principal)
df_web = tablas[0]
print("\n▶ Primeras 5 filas de la tabla web:")
display(df_web.head())
```

```python
# Limpieza y estructuración de datos para los tres datasets

print("=== Limpieza y estructuración de datos ===\n")

# --- Dataset CSV (Titanic) ---
print(">>> Dataset CSV (Titanic)")

# Mostrar valores nulos por columna
print("Valores nulos antes de limpiar:")
print(df_csv.isnull().sum())

# Decidir qué hacer con valores nulos:
# Por ejemplo, si existe 'Age' imputamos con la media (no eliminar filas)
if 'Age' in df_csv.columns:
    media_edad = df_csv['Age'].mean()
    df_csv['Age'] = df_csv['Age'].fillna(media_edad)
    print(f"Se imputaron valores nulos en 'Age' con la media: {media_edad:.2f}")

# Si existen columnas críticas con nulos (ejemplo 'Embarked'), eliminamos filas con nulos
if 'Embarked' in df_csv.columns:
    filas_antes = len(df_csv)
    df_csv = df_csv.dropna(subset=['Embarked'])
    filas_despues = len(df_csv)
    print(f"Se eliminaron {filas_antes - filas_despues} filas con valores nulos en 'Embarked'")

# Eliminar filas duplicadas si existen
duplicados_csv = df_csv.duplicated().sum()
print(f"Filas duplicadas encontradas: {duplicados_csv}")
df_csv = df_csv.drop_duplicates()

# Verificar tipos de datos
print("Tipos de datos después de limpieza:")
print(df_csv.dtypes)
print("\n")

# --- Dataset Excel (Liga Española) ---
print(">>> Dataset Excel (Liga Española)")

# Mostrar valores nulos
print("Valores nulos antes de limpiar:")
print(df_excel.isnull().sum())

# Suponemos que no hay nulos críticos, solo eliminamos duplicados
duplicados_excel = df_excel.duplicated().sum()
print(f"Filas duplicadas encontradas: {duplicados_excel}")
df_excel = df_excel.drop_duplicates()

# Verificar tipos de datos
print("Tipos de datos después de limpieza:")
print(df_excel.dtypes)
print("\n")

# --- Dataset Web (Población Mundial) ---
print(">>> Dataset Web (Población Mundial)")

# Mostrar valores nulos
print("Valores nulos antes de limpiar:")
print(df_web.isnull().sum())

# Convertir 'Population' a numérico, limpiando comas si las hay
df_web['Population'] = df_web['Population'].astype(str).str.replace(',', '', regex=False)
df_web['Population'] = pd.to_numeric(df_web['Population'], errors='coerce')

# Imputar valores nulos en 'Population' con la media
media_poblacion = df_web['Population'].mean()
df_web['Population'] = df_web['Population'].fillna(media_poblacion)
print(f"Se imputaron valores nulos en 'Population' con la media: {media_poblacion:.0f}")

# Eliminar filas duplicadas si existen
duplicados_web = df_web.duplicated().sum()
print(f"Filas duplicadas encontradas: {duplicados_web}")
df_web = df_web.drop_duplicates()

# Verificar tipos de datos
print("Tipos de datos después de limpieza:")
print(df_web.dtypes)
print("\n")
```

```python
print("=== Transformación y optimización de datos ===\n")

# --- Dataset CSV (Titanic) ---
print(">>> Dataset CSV (Titanic)")

# Selección de columnas relevantes para análisis (si existen)
# Adaptamos a las columnas disponibles, aquí el ejemplo más común:
columnas_titanic = ['PassengerId', 'Survived', 'Pclass', 'Sex', 'Age', 'Fare']
# Solo seleccionamos columnas que realmente existan en el DataFrame (por si algunas faltan)
columnas_titanic = [col for col in columnas_titanic if col in df_csv.columns]
df_csv_reducido = df_csv[columnas_titanic]

# Renombrar columnas para mejor legibilidad
df_csv_reducido = df_csv_reducido.rename(columns={
    'PassengerId': 'ID_Pasajero',
    'Survived': 'Sobrevivió',
    'Pclass': 'Clase_Pasajero',
    'Sex': 'Sexo',
    'Age': 'Edad',
    'Fare': 'Tarifa'
})

# Ordenar por ID de pasajero
df_csv_reducido = df_csv_reducido.sort_values(by='ID_Pasajero')

print("DataFrame Titanic transformado:")
display(df_csv_reducido.head())
print("\n")


# --- Dataset Excel (Liga Española) ---
print(">>> Dataset Excel (Liga Española)")

# Selección de columnas relevantes
columnas_liga = ['Date', 'Home Team', 'Away Team', 'Home Score', 'Away Score', 'Result']
columnas_liga = [col for col in columnas_liga if col in df_excel.columns]
df_excel_reducido = df_excel[columnas_liga]

# Renombrar columnas para claridad
df_excel_reducido = df_excel_reducido.rename(columns={
    'Date': 'Fecha',
    'Home Team': 'Equipo_Local',
    'Away Team': 'Equipo_Visitante',
    'Home Score': 'Goles_Local',
    'Away Score': 'Goles_Visitante',
    'Result': 'Resultado'
})

# Ordenar por fecha
df_excel_reducido = df_excel_reducido.sort_values(by='Fecha')

print("DataFrame Liga Española transformado:")
display(df_excel_reducido.head())
print("\n")


# --- Dataset Web (Población Mundial) ---
print(">>> Dataset Web (Población Mundial)")

# Selección de columnas relevantes
columnas_web = ['Country', 'Population', 'World %', 'Date']
columnas_web = [col for col in columnas_web if col in df_web.columns]
df_web_reducido = df_web[columnas_web]

# Renombrar columnas para mejor legibilidad
df_web_reducido = df_web_reducido.rename(columns={
    'Country': 'País',
    'Population': 'Población',
    'World %': 'Porcentaje_Mundial',
    'Date': 'Fecha'
})

# Ordenar por población de mayor a menor
df_web_reducido = df_web_reducido.sort_values(by='Población', ascending=False)

print("DataFrame Población Mundial transformado:")
display(df_web_reducido.head())
print("\n")
```

```python
import os
from google.colab import files

print("=== Exportación de datos ===\n")

# --- Exportar Dataset CSV (Titanic) ---
csv_path_csv = "Titanic_Limpio.csv"
excel_path_csv = "Titanic_Limpio.xlsx"

df_csv_reducido.to_csv(csv_path_csv, index=False)
df_csv_reducido.to_excel(excel_path_csv, index=False)

print(f"Archivos Titanic exportados: {csv_path_csv} y {excel_path_csv}")

# --- Exportar Dataset Excel (Liga Española) ---
csv_path_excel = "LigaEspanola_Limpio.csv"
excel_path_excel = "LigaEspanola_Limpio.xlsx"

df_excel_reducido.to_csv(csv_path_excel, index=False)
df_excel_reducido.to_excel(excel_path_excel, index=False)

print(f"Archivos Liga Española exportados: {csv_path_excel} y {excel_path_excel}")

# --- Exportar Dataset Web (Población Mundial) ---
csv_path_web = "PoblacionMundial_Limpio.csv"
excel_path_web = "PoblacionMundial_Limpio.xlsx"

df_web_reducido.to_csv(csv_path_web, index=False)
df_web_reducido.to_excel(excel_path_web, index=False)

print(f"Archivos Población Mundial exportados: {csv_path_web} y {excel_path_web}")

# Listar archivos en el directorio actual para ver los generados
print("\nArchivos en el directorio actual tras exportar:")
print(os.listdir())

# Descargar archivos
print("\nIniciando descarga de archivos...")

files.download(csv_path_csv)
files.download(excel_path_csv)

files.download(csv_path_excel)
files.download(excel_path_excel)

files.download(csv_path_web)
files.download(excel_path_web)

print("\n¡Descarga completada!")
```

Se renombran columnas para mejorar legibilidad y facilitar el análisis.

3. Transformación y optimización

Seleccionamos solo las columnas relevantes para cada dataset, lo que mejora rendimiento y claridad.

Se ordenan los datos según una columna clave para facilitar consultas cronológicas o por tamaño.

Los nombres de columnas se adaptan a un idioma común y formato amigable.

4. Exportación

Se guardan los datasets limpios en archivos CSV y Excel para compartir o seguir trabajando con ellos.

No se incluye índice en los archivos para evitar columnas innecesarias al abrirlos en Excel u otras herramientas.
